```python
class Solution(object):
    def reconstructQueue(self, people):
        """
        :type people: List[List[int]]
        :rtype: List[List[int]]
        """
        import collections
        qdict = collections.defaultdict(list)
        
        l = len(people)
        for ppl in people:
            qdict[ppl[1]].append(ppl)
        for k in qdict.keys():
            qdict[k] = sorted(qdict[k], key= lambda p : p[0])
        print qdict
        # qdict[0].pop(0)
        # print qdict
        # print sorted(qdict.keys())
        tlist = qdict[0]
        print tlist
        l -= len(tlist)
        while l:
            for k in sorted(qdict.keys()):
                if k == 0:
                    continue
                if qdict[k]:
                    # print qdict[k]
                    hdict = {}
                    for h in set([q[0] for q in qdict[k]]):
                        hdict[h] = 0
                    tl = len(tlist)
                    i = 0
                    while i < tl:
                        if qdict[k] == []:
                            i = len(tlist)
                            continue
                            
                        th = tlist[i][0]
                        for hk in hdict.keys():
                            if hk <= th:
                                hdict[hk] += 1
                        if hdict[qdict[k][0][0]] == k:
                            tlist = tlist[:i+1] + [qdict[k][0]] + tlist[i+1:]
                            qdict[k].pop(0)
                            l -= 1
                            tl += 1
                            
                            print tlist
                        # print k,':',qdict[k]
                        i += 1
                else:
                    continue
        return tlist
        
                    
                    
```
